```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
features= [' Subflow Fwd Bytes', 'Total Length of Fwd Packets', ' Fwd Packet Length Std', ' min_seg_size_forward',  ' Flow IAT Min']
X = df_.loc[:, features].values
Y = df_.loc[:,[' Label']].values

x = StandardScaler().fit_transform(X)
```

```python
pca = PCA(.95)
principal_components = pca.fit_transform(x)
pca_df = pd.DataFrame(data = principal_components)
pca_df.head()
```

```python
df_final = pd.concat([pca_df, df_[[' Label']]], axis = 1)
df_final.head()
```
